```python
import os
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt

diretorio = 'imagens'

for arquivo in os.listdir(diretorio):
  dir = f'{diretorio}/{arquivo}'
  # Verifica se o arquivo é uma imagem antes de processá-lo
  if arquivo.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
    img = cv2.imread(dir)
    # Verifica se a imagem foi carregada corretamente
    if img is not None:
      img = cv2.resize(img,(640,480))
      img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

      resultado = DeepFace.analyze(img,actions=['emotion'])
      emocoes = resultado[0]['emotion']
      x = list(emocoes.values())
      y = list(emocoes.keys())

      fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,5))

      ax2.barh(y,x)
      ax2.tick_params(labelsize=10)
      ax1.imshow(img)
    else:
      print(f"Aviso: Não foi possível carregar a imagem em {dir}")
  else:
    print(f"Aviso: Ignorando arquivo {arquivo}, pois não é uma imagem.")
```
